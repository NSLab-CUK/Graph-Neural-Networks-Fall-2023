### Exercise 1. 

Given an undirected graph has 8 nodes: 0,1,2,3,4,5,6,7

and the edge list: (0, 1), (0, 2), (0, 3), (0, 5), (1, 3), (1, 6), (3, 4), (4, 5), (4, 7)

Question:
- Draw the graph.
- Count number of nodes and edges in the graph.
- List all of nodes, edges in the graph.

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt

G = nx.Graph()

G.add_edges_from([(0, 1),
                   (0, 2),
                   (0, 3),
                   (0, 5),
                   (1, 3),
                   (1, 6),
                   (3, 4),
                   (4, 5),
                   (4, 7)])

print("# nodes: ", int(G.number_of_nodes()))
print("# edges: ", int(G.number_of_edges()))
print("List of nodes: ", list(G.nodes()))
print("List of edges: ", list(G.edges(data = True)))
```

```python
# draw the network G
plt.figure(figsize =(6, 6))
nx.draw_networkx(G, with_labels = True, node_color ='green')
```
